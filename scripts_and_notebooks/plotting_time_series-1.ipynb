```python
import numpy as np
import pandas as pd
import scipy.io as sio
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
path = "../Injection-Heights/"
# Old Heights 2 August 2019
oldh2 = sio.loadmat(path+"injection-heights_2-Aug-2019.mat")
# New Heights 2 August 2019
newh2 = sio.loadmat(path+"new-injection-heights_2-Aug-2019.mat")

# Old Heights 3 August 2019
oldh3 = sio.loadmat(path+"injection-heights_3-Aug-2019.mat")
# New Heights 3 August 2019
newh3 = sio.loadmat(path+"new-injection-heights_3-Aug-2019.mat")
```

```python
alltimes = np.array([])

times2 = np.array(oldh2['time'][0])

times3 = np.array(oldh3['time'][0])
tdelta = np.full(len(times3), 24*60*60)

times3 = times3 + tdelta

# Appending all the times to one array
alltimes = np.append(alltimes, times2)
alltimes = np.append(alltimes, times3)

# times to hours
alltimes = alltimes/3600
```

```python
alltimes
```

```python
"""
Williams Flats Fire first reported 3:23 AM PST https://wildfiretoday.com/tag/williams-flats-fire/
"""

start_time = 10 + 23/60 # 10:23 am UTC time

start_time
```

```python
# Extracting the heights for each time the heights
# were retrieved

oldheights = np.array([]) 
newheights = np.array([])

# 2 August 2019
for t in range(len(times2)):
    # Retrieving the max heights for both old and new heights
    oldmax = np.max(oldh2['smoke_injection_height'][:, :, t]) 
    newmax = np.max(newh2['smoke_injection_height'][:, :, t]) 
    
    if oldmax < 0:
        oldheights = np.append(oldheights, np.NaN)
    else:
        oldheights = np.append(oldheights, oldmax)
        
    if newmax < 0:
        newheights = np.append(newheights, np.NaN)
    else:
        newheights = np.append(newheights, newmax)
        
# 3 August 2019
for t in range(len(times3)):
    # Retrieving the max heights for both old and new heights
    oldmax = np.max(oldh3['smoke_injection_height'][:, :, t])
    newmax = np.max(newh3['smoke_injection_height'][:, :, t])
    
    if oldmax < 0:
        oldheights = np.append(oldheights, np.NaN)
    else:
        oldheights = np.append(oldheights, oldmax)
        
    if newmax < 0:
        newheights = np.append(newheights, np.NaN)
    else:
        newheights = np.append(newheights, newmax)
```

```python
plt.figure(figsize=(20,10))
plt.grid()
plt.plot(alltimes, oldheights, marker="x", color="red", label="Ref $\geq$ 10 dBZ")
plt.plot(alltimes, newheights, marker="v", color="green", label="Ref $\geq$ 10 dBZ, 0.2 < C.C. < 0.9")
plt.axvline(start_time, color="blue", linestyle="--")
plt.text(10.45, 10800, "Williams Flats Fire first reported, 10:23am UTC", fontsize=13, color="blue")
plt.title("2 August 2019 UTC Injection Heights", fontsize=23)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.xlim([7, 31])
plt.xlabel("Time [hours] after 00:00 UTC", fontsize=20)
plt.ylabel("Injection Height [meters above sea level]", fontsize=20)
plt.legend(fontsize=13, loc="best")

plt.savefig("../Visualizations/comparing-algorithms-2aug2019.png")
```
